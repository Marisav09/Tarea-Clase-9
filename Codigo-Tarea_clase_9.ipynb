## Descripción del Problema:
Escribir un programa en Python que administre una lista de tareas. El programa debe permitir al usuario agregar, eliminar y listar tareas. Cada tarea tendrá un nombre, una descripción y una prioridad (alta, media, baja). Además, habrá diferentes tipos de tareas: TareaGeneral y TareaConFecha, cada una con características adicionales.

### Requisitos del Código:

1. El programa debe permitir agregar una nueva tarea.
2. El programa debe permitir eliminar una tarea existente.
3. El programa debe permitir listar todas las tareas con sus detalles.
4. El programa debe estar diseñado utilizando POO e incluir herencia.

## Lista de tareas:
### Creacion de Tareas Generales y Tareas con Fecha

```python
class TareasGenerales:
    def __init__(self, nombre,descripcion,prioridad):
        self.nombre=nombre
        self.descripcion=descripcion
        self.prioridad=prioridad
    def info_tarea(self):
        print(f" Nombre: {self.nombre}\n Descripcion: {self.descripcion}\n Prioridad: {self.prioridad}\n")

class TareaConFecha(TareasGenerales):
    def __init__(self, nombre, descripcion, prioridad,fecha):
        super().__init__(nombre,descripcion,prioridad)
        self.fecha=fecha
    def info_tarea_fecha(self):
        super().info_tarea()
        print(f"Fecha{self.fecha}")
```

### Agregar una nueva tarea.

```python
lista_tareas = []
def agregar_tarea(tarea):
    lista_tareas.append(tarea)
    print(f"La tarea fue agregada")
 
tarea1 = TareasGenerales("Comprar comestibles", "Ir al supermercado y comprar víveres", "Alta")
agregar_tarea(tarea1)
tarea1.info_tarea()
```

```python
from datetime import datetime

class Tarea:
    def __init__(self, nombre, descripcion, prioridad):
        self.nombre = nombre
        self.descripcion = descripcion
        self.prioridad = prioridad

    def __str__(self):
        return f"Nombre: {self.nombre}\nDescripción: {self.descripcion}\nPrioridad: {self.prioridad}"

class TareaConFecha(Tarea):
    def __init__(self, nombre, descripcion, prioridad, fecha):
        super().__init__(nombre, descripcion, prioridad)
        self.fecha = fecha

    def __str__(self):
        return super().__str__() + f"\nFecha: {self.fecha.strftime('%Y-%m-%d')}"

class GestorDeTareas:
    def __init__(self):
        self.tareas = []

    def agregar_tarea(self, tarea):
        self.tareas.append(tarea)
        print("Tarea agregada exitosamente.")

    def eliminar_tarea(self, nombre):
        for tarea in self.tareas:
            if tarea.nombre == nombre:
                self.tareas.remove(tarea)
                print("Tarea eliminada exitosamente.")
                return
        print("No se encontró la tarea.")

    def listar_tareas(self):
        if not self.tareas:
            print("No hay tareas en la lista.")
        else:
            print("Lista de Tareas:")
            for tarea in self.tareas:
                print(tarea)
                print()

# Ejemplo de uso
gestor = GestorDeTareas()

tarea1 = Tarea("Comprar comestibles", "Ir al supermercado y comprar víveres", "Alta")
tarea2 = TareaConFecha("Llamar a mamá", "Felicitarla por su cumpleaños", "Media", datetime(2024, 6, 20))

gestor.agregar_tarea(tarea1)
gestor.agregar_tarea(tarea2)

gestor.listar_tareas()

gestor.eliminar_tarea("Comprar comestibles")

gestor.listar_tareas()
```

```python
class TareasGenerales:
    def __init__(self, nombre,descripcion,prioridad):
        self.nombre=nombre
        self.descripcion=descripcion
        self.prioridad=prioridad
    def info_tarea(self):
        print(f" Nombre: {self.nombre}\n Descripcion: {self.descripcion}\n Prioridad: {self.prioridad}\n")

class TareaConFecha(TareasGenerales):
    def __init__(self, nombre, descripcion, prioridad,fecha):
        super().__init__(nombre,descripcion,prioridad)
        self.fecha=fecha
    def info_tarea_fecha(self):
        super().info_tarea()
        print(f"Fecha{self.fecha}")

lista_tareas = []
def agregar_tarea():
    nombre = input("Ingrese el nombre de la tarea: ")
    descripcion = input("Ingrese la descripción de la tarea: ")
    prioridad = input("Ingrese la prioridad de la tarea (Alta media o baja): ")
    if prioridad.lower() != "alta" or prioridad.lower() != "media" or prioridad.lower() != "baja":
        prioridad=print ("Elija la una de las tres opciones por favor (Alta Medio a o baja)") 
    else: 
        pass
    tiene_fecha = input("¿Desea agregar una fecha a la tarea? (si/no): ")
    if tiene_fecha.lower() == "si":
        fecha = input("Ingrese la fecha de la tarea (DD-MM-YYYY): ")
        tarea = TareaConFecha(nombre, descripcion, prioridad, fecha)
    else:
        tarea = TareasGenerales(nombre, descripcion, prioridad)
    lista_tareas.append(tarea)
    print(f"La tarea '{tarea.nombre}' fue agregada.")
 
def eliminar_tarea(nombre):
    for tarea in lista_tareas:
        if tarea.nombre == nombre:
            lista_tareas.remove(tarea)
            print("Tarea eliminada exitosamente.")
            return
    print("La tarea no existe en la lista.")

def listar_tareas():
    if not lista_tareas:
        print("No hay tareas.")
    else:
        print("\nTareas sin fecha:")
        for tarea in lista_tareas:
            if not isinstance(tarea, TareaConFecha):
                tarea.info_tarea()
                print()

        print("Tareas con fecha:")
        for tarea in lista_tareas:
            if isinstance(tarea, TareaConFecha):
                tarea.info_tarea_fecha()
                print()


while True:
    print()
    opc=input("Bienvenido a la funcion TAREAS.\nPor favor elija la opcion (numerica) que desea realizar: \n1= Listar tareas actuales \n2= Agregar tareas \n3= Eliminar tareas \n4=Salir")
    if opc == "1":
        listar_tareas()
    elif opc == "2":
        agregar_tarea()
        
    elif opc == "3":
        nombre=input("Ingrese el nombre de la tarea que desea eliminar")
        eliminar_tarea(nombre)
    elif opc == "4":
        print("¡Gracias por su visita!")
        break
    else:
        print("Opción inválida. Intente de nuevo.")
```
